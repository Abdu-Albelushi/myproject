```
import ipywidgets as widgets
from IPython.display import display, clear_output

to_do_list= []

welcome_label= widgets.HTML(value="<h2> Welcome to Your Visual To-Do List!</h2>")
instructions= widgets.HTML(value="Type a task below and click 'add'. View or remove tasks any time.")
task_input= widgets.Text(Placeholder='Enter a task...')
add_button= widgets.Button(description='Add Task', button_style='success')
view_button= widgets.Button(description='View List', button_style='info')
remove_button= widgets.Button(description='Remove Task', button_style='danger')
output_area= widgets.Output()

def add_task(b):
  task= task_input.value.strip()
  if task:
    to_do_list.append(task)
    output_area.clear_output()
    with output_area:
      print(f"Added: {task}")
  else:
    output_area.clear_output()
    with output_area:
      print("Please enter a task.")

task_input.value= ""

def view_list(b):
  output_area.clear_output()
  with output_area:
    if to_do_list:
      print("Your To-Do List:")
      for idx, task in enumerate(to_do_list, start=1):
        print(f"{idx}. {task}")
    else:
      print("Your To-Do List is empty.")

def remove_task(b):
  task= task_input.value.strip()
  output_area.clear_output()
  if task in to_do_list:
    to_do_list.remove(task)
    with output_area:
      print(f"Removed: {task}")
  else:
    with output_area:
      print("That task isn't on your list!")
  task_input.value=""

add_button.on_click(add_task)
view_button.on_click(view_list)
remove_button.on_click(remove_task)

display(welcome_label, instructions, task_input, widgets.HBox([add_button, view_button, remove_button]), output_area)


```
